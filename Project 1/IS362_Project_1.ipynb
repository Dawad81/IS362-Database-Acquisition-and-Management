I created a .csv file to represent the data named `Alaska_and AM_West_Flights.csv` we will now import that file into a data frame.

```python
import pandas as pd
flight_data = pd.read_csv('Alaska_and AM_West_Flights.csv', index_col='Airline')

flight_data
```

Now we will neeed the precentage of delayed flights by airline at each airport. to do this we will add the `Total_Flight_Time` colume that is the sum or the `OnTime` `+` `Delayed` coulums.
we can then use the new `Total_Flight_Time` colume to figure out the `Percentage_of_Flights_Delayed` colume by dividing the `Delayed` by the `Total_Flight_Time` colume
 we will make the `Percentage_of_Flights_Delayed` colume easier to understand by formating it to show %

```python
flightDataEdits = flight_data.copy()
flightDataEdits.insert(3, 'Total_Flight_Time', flight_data.OnTime + flight_data.Delayed)
flightDataEdits.insert(4, 'Percentage_of_Flights_Delayed', (100 * flight_data.Delayed / flightDataEdits.Total_Flight_Time))
pd.options.display.float_format = '{:.1f}%'.format

flightDataEdits
```

Below we can now see the percentage of delayed flights and their corisponding airports and airline

```python
flightDataEdits[['Airport', 'Percentage_of_Flights_Delayed']]
```

To make life a bit easier we can create a function to use to serch the data frame. we will create the `airportDelayCompare()` function to do this for us.

```python
def airportDelayCompare(airport):
    delays = flightDataEdits.loc[flightDataEdits['Airport']== airport]
    return delays[['Airport', 'Percentage_of_Flights_Delayed']]
```

Now we will reset our index to add `Percentage_of_Flights_Delayed` to the index. we will first make a copy of `flightDataEdits` named `resetindex`

```python
resetindex=flightDataEdits.copy()
resetindex.set_index('Percentage_of_Flights_Delayed', append=True)
```

We can now use `resetindex` to group airlines percentages together to get Total OnTime perecentage for each airline

```python
AirlineComparison = resetindex
AirlineComparison = resetindex.groupby(["Airline"])["OnTime", "Delayed"].sum()
AirlineComparison["Percentage_of_Flights_OnTime"] = 100 * AirlineComparison["OnTime"] / (AirlineComparison["OnTime"] + AirlineComparison["Delayed"])
AirlineComparison
```

with a few edits we can also get the total Delayed percentages of each air line as well

```python
AirlineComparison = resetindex
AirlineComparison = resetindex.groupby(["Airline"])["OnTime", "Delayed"].sum()
AirlineComparison["Percentage_of_Flights_Delayed"] = 100 * AirlineComparison["Delayed"] / (AirlineComparison["OnTime"] + AirlineComparison["Delayed"])
AirlineComparison
```
