```python
import pandas as pd

flight_data = pd.read_csv('Alaska_and AM_West_Flights.csv', index_col='Airline')

flight_data
```

```python
flightDataEdits = flight_data.copy()
flightDataEdits.insert(3, 'Total_Flight_Time', flight_data.OnTime + flight_data.Delayed)
flightDataEdits.insert(4, 'Percentage_of_Flights_Delayed', (flight_data.Delayed / flightDataEdits.Total_Flight_Time))
pd.options.display.float_format = '{:.2f}%'.format

flightDataEdits
```

```python
def airportDelayCompare(airport):
    #need to convert airport into string to work
    delays = flightDataEdits.loc[flightDataEdits['Airport']== airport]
    return delays

#need to work on this function
```
