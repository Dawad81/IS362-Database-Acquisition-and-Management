```python
salery_survey = pd.read_csv('Ask A Manager Salary Survey 2019 (Responses) - Form Responses 1.csv', header=1)
salery_survey
```

```python
salery_survey.isnull().sum()
```

```python
column_names = {'How old are you?':'age',
                'What industry do you work in?':'industry',
                'Job title':'job',
                'What is your annual salary?':'salary',
                'Please indicate the currency':'currency',
                'Where are you located? (City/state/country)':'city/state/country',
                'How many years of post-college professional work experience do you have?':'years_exsperience'
               }
tidey_salery_survey = (salery_survey.rename(columns=column_names)
                       [['age','industry','job', 'salary','currency','city/state/country','years_exsperience']]
                       .rename_axis('entry#').sort_index())
tidey_salery_survey
```

```python
tidey_salery_survey.isnull().sum()
```

```python
tidey_salery_survey_copy = tidey_salery_survey.copy()
tidey_salery_survey_copy
```

```python
tidey_salery_survey_copy.fillna(value='UNKNOWN', inplace=True)
tidey_salery_survey_copy
```

```python
tidey_salery_survey_copy.isnull().sum()
```

```python
tidey_salery_survey_copy['age'].value_counts()
```

```python
Top_salerys = tidey_salery_survey_copy.sort_values(by='salary', ascending = False).head(100)
Top_salerys
```

```python
Top_salerys['currency'].value_counts()
```

```python
tidey_salery_survey_copy['currency'].value_counts()
```

in the us what age group makes the most money

```python
us_salaries = tidey_salery_survey_copy.loc[tidey_salery_survey_copy['currency']== 'USD']
us_salaries
```

```python
us_salaries.salary = pd.to_numeric(us_salaries['salary'], errors='coerce')
```

```python
us_salaries
```

```python
us_salaries_int = us_salaries.dropna()
us_salaries_int
```

```python
us_salaries_int.sort_values(by='salary', ascending = False)
```

```python
us_salaries_int['age'].value_counts()
```

```python
import numpy as np
```

```python
table= pd.pivot_table(us_salaries_int, values='salary', index='age', aggfunc={'salary':[min, max,np.mean]})
#we format flotes to make the salaries more readable
pd.options.display.float_format = '{:.2f}'.format
table
```

```python
table2= pd.pivot_table(us_salaries_int, values='salary', index='years_exsperience', aggfunc={'salary':[min, max,np.mean]})
table2
```
