# We will use this `cars` dataset to find out which `Origin` produces the most fule efficiant cars in the list, And out of those results, what most fule efficient car in that `Origin` has the most `Horsepower`.

***
Now in order to accomlish this, we will remove some of the columns in the `cars` data set, This will give us a cleaner data set that we will name `tidey_cars`. `tidey_cars` will only consit of the `car_id`, `Car`, `MPG`, `Horspower`, `Cylinders`, and `Origin`.

```python
column_names = {'Car':'Car', 'MPG':'MPG', 'Horsepower':'Horsepower', 'Cylinders':'Cylinders','Origin':'Origin'}
tidey_cars = (cars.rename(columns=column_names)[['Car', 'MPG', 'Horsepower', 'Cylinders','Origin']].rename_axis(['car_id']).sort_index())
tidey_cars
```

Now that we have a `tidey_cars` data set, to find the most fule efficant cars we will take the top 10 results and sort them by fule efficency. To do this we will sort them in descending order by `MPG`. We will name this list `most_fule_efficiant_cars`.

```python
most_fule_efficiant_cars = tidey_cars.sort_values(by = 'MPG', ascending=False).head(10)
most_fule_efficiant_cars
```

We can see our first result is a V8 with 9.0 MPG which is definitly some bad data. However the other 9 results seem to be on track with an `MPG` in the 40+ range, and `Europe` seems to have a dominit hold.

To be sure we will do a quick count of the `Origins` in `most_fule_efficiant_cars`.

```python
most_fule_efficiant_cars['Origin'].value_counts()
```

Yep, loos like `Europe` has 6 cars in the top 10 most fule efficient cars in the data set. Let clean this up further and return just those 6 cars in to the variable `europes_most_fule_efficiant_cars`

```python
europes_most_fule_efficiant_cars = most_fule_efficiant_cars.loc[most_fule_efficiant_cars['Origin']=='Europe']
europes_most_fule_efficiant_cars
```

Now that we have this list of `europes_most_fule_efficiant_cars` lets find the one with the most horspower.

```python
europes_most_fule_efficiant_cars.sort_values(by = 'Horsepower',ascending=False).head(1)
```

Looks like `car_id`:317 is the winner. The `Volkswagen Rabbit` with and `MPG` of 41.5 has the most `Horsepower` with 76 horses.

## If you want the most fule efficient car, buy a European car, becuse they makes 6 of the top 10 most fule efficent cars in this list. And if you want one with the most Horspower the `Volkswagen Rabbit` is your pick with 76 Horses.

We will import the data from the `Ask A Manager Salary Survey 2019 (Responses) - Form Responses 1.csv` file into the `salery_survey` variable.

```python
salery_survey = pd.read_csv('Ask A Manager Salary Survey 2019 (Responses) - Form Responses 1.csv', header=1)
salery_survey
```

Lets take deeper look at this file for `NaN`'s

```python
salery_survey.isnull().sum()
```

***
Looks like were going to have to clean up these column names, and fill in these `NaN` values. So lets do that and create the `tidey_salery_survey`.

```python
column_names = {'How old are you?':'age',
                'What industry do you work in?':'industry',
                'Job title':'job',
                'What is your annual salary?':'salary',
                'Please indicate the currency':'currency',
                'Where are you located? (City/state/country)':'city/state/country',
                'How many years of post-college professional work experience do you have?':'years_exsperience'
               }
tidey_salery_survey = (salery_survey.rename(columns=column_names)
                       [['age','industry','job', 'salary','currency','city/state/country','years_exsperience']]
                       .rename_axis('entry#').sort_index())
tidey_salery_survey
```

Lets check our ` Nan`s and in the process we should make sure our column names changed.

```python
tidey_salery_survey.isnull().sum()
```

Great that worked. Since we are dealling with a .csv file that we downloaded and dont want to accidently change values in the .csv file lets make a new variable for safty called `tidey_salery_survey_copy`. well work with this file to make changes from here on out.

```python
tidey_salery_survey_copy = tidey_salery_survey.copy()
tidey_salery_survey_copy
```

Lets fill the `Nan` values with the `str` `'UNKNOWN'`.

```python
tidey_salery_survey_copy.fillna(value='UNKNOWN', inplace=True)
tidey_salery_survey_copy
```

Lets confirm we no longer have any `NaN` Values.

```python
tidey_salery_survey_copy.isnull().sum()
```

Great we have no `Nan`s.

# Now we will try to use this data to find which age group in the US has the highest average salary.

Lets take a look at our age Groups.

```python
tidey_salery_survey_copy['age'].value_counts()
```

Lets pull out all the data that is in USD curency, giving out US results.

```python
us_salaries = tidey_salery_survey_copy.loc[tidey_salery_survey_copy['currency']== 'USD']
us_salaries
```

***
We will have to make sure this `salary` data is cleaned and converted to Integers so we can do some math on it later.

```python
us_salaries.salary = pd.to_numeric(us_salaries['salary'], errors='coerce')
```

Lets take a look at what `us_salaries` look like now.

```python
us_salaries
```

Well looks like in the conversion we had some bad values we couldnt conver that resulted in `NAN` values lets get rid of those well call this new dataset  `us_salaries_int`.

```python
us_salaries_int = us_salaries.dropna()
us_salaries_int
```

Lets sort this from Highest to lowes `salary`.

```python
us_salaries_int.sort_values(by='salary', ascending = False)
```

Were gona have to use `numpy` to do some math, so lets import that to help us in creating our `pivot_table`

```python
import numpy as np
```

Lets creat a pivot table  with the`mean` saleries for each age group.

```python
table= pd.pivot_table(us_salaries_int, values='salary', index='age', aggfunc={'salary':np.mean})
#we format flotes to make the salaries more readable
pd.options.display.float_format = '{:.2f}'.format
table
```

## Looks like the `55-64` age group has the highest average salary of `$148,665.80`

***


Lets do another pivot table to see what numer of years exsperience makes the most money in the survey.

```python
table2= pd.pivot_table(us_salaries_int, values='salary', index='years_exsperience', aggfunc={'salary':np.mean})
table2
```

### Looks like the highest average salery belongs to those with 21-30 years of exsperience.

# Third Dataset: Film data


This was the data set used in Taha's  week 5 discussion board. This data has about 1600 movies with properties such as length, main actor and actress, director and popularity.

Retrived from:  [Project Datasets Site](https://perso.telecom-paristech.fr/eagan/class/igr204/datasets)

## We will use this data to list the `Drama` movies that came out in the `1980'S`.

First we will import this dataset from the website in to the variable `films`.

```python
films = pd.read_csv('https://perso.telecom-paristech.fr/eagan/class/igr204/data/film.csv', sep=';')
films
```

As we can see row 0 is a guide, to inform us of how the data in the corresponding column should be represented. We do not need this row so we will remove it just like we did in out first dataset.

```python
films.drop(films.index[0], inplace=True)
films
```

Lets take a look at our counts for type of movies using the `Subject` column

```python
films.Subject.value_counts()
```

Now lets make sure all the rows that should be integers, are integers. So we will convert `Year`, `Length`, and `Popularity` to integers.

```python
films.Year = pd.to_numeric(films['Year'], errors='coerce')
films.Length = pd.to_numeric(films['Length'], errors='coerce')
films.Popularity = pd.to_numeric(films['Popularity'], errors='coerce')
```

Now we can query our `films` dataset for `Drama` films from the 1980's (so all years between `1980` and `1989`) and we will sort them according to `popularity`.

```python
films.loc[(films['Year'] >= 1980) & (films['Year']<= 1989) & (films['Subject']=='Drama')].sort_values(by='Popularity', ascending = False)
```

## SO it looks like the top `Drama` film from the 1980's is actually a three way tie between `Outh of Africa` in 1985, `Five Corners` in 1988, and `The Ballad of Narayama` in 1982.
