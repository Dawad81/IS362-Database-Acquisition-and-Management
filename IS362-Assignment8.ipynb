## 4. Some of the horsepower data is marked as missing with a ‘?’. You need to replace this ‘?’ with a pandas-appropriate missing value, then convert the column to numeric.

Lets find the rows with a '`?`' in the `horsepower` column.

```python
auto_mpg.loc[auto_mpg['horsepower']=='?']
```

Now lets change those `?` into `Nan`'s and then serch again to see if we have any `?` in the `horsepower` column

```python
auto_mpg = auto_mpg.replace('?', np.nan)

auto_mpg.loc[auto_mpg['horsepower']=='?']
```

Great we got no results so all the `?` have been removed. But to be sure lets serch for the new `NaN`'s and see if they match what we got befor when serching for `?`.

```python
auto_mpg.loc[auto_mpg.horsepower.isnull()]
```

We can see that the `?` have all been switch to NaN and the results of our serches are identical.
***


Now lets chage all the rows in the `horsepower`column to numeric results.

```python
auto_mpg['horsepower'] = pd.to_numeric(auto_mpg['horsepower'],errors='coerce')
auto_mpg
```

Lets check the `horsepower` columns data type.

```python
auto_mpg.horsepower.dtypes
```

## 5. The origin column values are 1, 2, and 3. Convert these values respectively to ‘USA’, ‘Asia’, and ‘Europe’


Here we will use a dict to replce all the values in the `origin` column with the coresponding values `1='USA', 2='Europe', 3='Asia' `

```python
auto_mpg = auto_mpg.replace({'origin' : { 1:'USA', 2:'Europe', 3:'Asia' }})
auto_mpg
```

## 6. Create a bar chart that shows the distribution for cylinders.


First we will create a variable `cylinders_bar_graph` that will count the values in the `cylinders` column.

```python
cylinders_bar_graph = auto_mpg['cylinders'].value_counts(dropna=False)
cylinders_bar_graph
```

Now we will plot the `cylinders_bar_graph` in to a bar chart

```python
cylinders_bar_graph.plot(kind='bar')
plt.title('Distrabution of Cylnders')
plt.xlabel('Cylinders')
plt.ylabel('Number of Cars')
```

## 7. Create a scatterplot that shows the relationship between horsepower and weight.

```python
sb.lmplot(data = auto_mpg, x = 'weight', y = 'horsepower', height = 5)
plt.title('Horsepower Vs. Weight')
plt.xlabel('Weight')
plt.ylabel('Horsepower')
```

Above we can see the realionships of `Horsepower vs Weight` of all of the cars in the auto_mpg dataframe.

## 8. Ask, then answer (using matplotlib and seaborn graphics capabilities) an interesting question about the distribution of one of the variables, or the relationship between two (or more!) of the variables in the “Auto MPG” dataset.

What if we can seperatly chart the `Horsepower vs Weight` and compare them by `origin` to get a better sence of whats going on.

```python
#This will give us Horspower Vs. Weight in the USA
sb.lmplot(data = auto_mpg[auto_mpg['origin']== 'USA'], x = 'weight', y = 'horsepower', height = 5)
plt.title('Horsepower Vs. Weight in the USA')
plt.axis([1400, 5000, 0, 250])
plt.xlabel('Weight')
plt.ylabel('Horsepower')


#This will give us Horspower Vs. Weight in Europe
sb.lmplot(data = auto_mpg[auto_mpg['origin']== 'Europe'], x = 'weight', y = 'horsepower', height = 5)
plt.title('Horsepower Vs. Weight in Europe')
plt.axis([1400, 5000, 0, 250])
plt.xlabel('Weight')
plt.ylabel('Horsepower')


#This will give us Horspower Vs. Weight in Asia
sb.lmplot(data = auto_mpg[auto_mpg['origin']== 'Asia'], x = 'weight', y = 'horsepower', height = 5)
plt.axis([1400, 5000, 0, 250])
plt.title('Horsepower Vs. Weight in Asia')
plt.xlabel('Weight')
plt.ylabel('Horsepower')
```
