```python
import pandas as pd
import numpy as np
import seaborn as sb
from matplotlib import pyplot as plt
%matplotlib inline


auto_mpg = pd.read_csv('Week8-DataFolder/auto-mpg.data', delim_whitespace = True)
auto_mpg
```

```python

auto_mpg_names = open(r'Week8-DataFolder/auto-mpg.names')
auto_mpg_names.readlines()
```

```python
auto_mpg_columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model year', 'origin', 'car name']
auto_mpg.columns = auto_mpg_columns
auto_mpg
```

```python
auto_mpg.loc[auto_mpg['horsepower']=='?']
```

```python
auto_mpg = auto_mpg.replace('?', np.nan)

auto_mpg.loc[auto_mpg['horsepower']=='?']
```

```python
auto_mpg.loc[auto_mpg.horsepower.isnull()]
```

```python
auto_mpg['horsepower'] = pd.to_numeric(auto_mpg['horsepower'],errors='coerce')
auto_mpg
```

```python
auto_mpg.horsepower.dtypes
```

```python
auto_mpg = auto_mpg.replace({'origin' : { 1:'USA', 2:'Europe', 3:'Asia' }})
auto_mpg
```

```python
cylinders_bar_graph = auto_mpg['cylinders'].value_counts(dropna=False)
cylinders_bar_graph
```

```python
cylinders_bar_graph.plot(kind='bar')
```

```python
cylinders_bar_graph2 = auto_mpg.groupby(['cylinders']).size()
cylinders_bar_graph2
```

```python
cylinders_bar_graph2.plot(kind='bar')
```

```python

sb.lmplot(data = auto_mpg, x = 'weight', y = 'horsepower', height = 5)
plt.title('Horsepower Vs. Weight')
plt.xlabel('Weight')
plt.ylabel('Horsepower')
```

```python
sb.lmplot(data = auto_mpg[auto_mpg['origin']== 'USA'], x = 'weight', y = 'horsepower', height = 5)
plt.title('Horsepower Vs. Weight in the USA')
plt.axis([1400, 5000, 0, 250])
plt.xlabel('Weight')
plt.ylabel('Horsepower')
```

```python
sb.lmplot(data = auto_mpg[auto_mpg['origin']== 'Europe'], x = 'weight', y = 'horsepower', height = 5)
plt.title('Horsepower Vs. Weight in Europe')
plt.axis([1400, 5000, 0, 250])
plt.xlabel('Weight')
plt.ylabel('Horsepower')
```

```python
sb.lmplot(data = auto_mpg[auto_mpg['origin']== 'Asia'], x = 'weight', y = 'horsepower', height = 5)
plt.axis([1400, 5000, 0, 250])
plt.title('Horsepower Vs. Weight in Asia')
plt.xlabel('Weight')
plt.ylabel('Horsepower')
```

```python
#asia=auto_mpg[auto_mpg['origin']== 'Asia']
#pd.options.display.mpl_style = 'default'
multiplot1 = plt.subplot2grid(shape=(3,1), loc=(0,0))
multiplot1.plot(kind='scatter',x='weight',y='horsepower')
plt.axis([1400, 5000, 0, 250])
plt.title('Horsepower Vs. Weight in the USA')
plt.xlabel('Weight')
plt.ylabel('Horsepower')

multiplot2 = plt.subplot2grid(shape=(3,1), loc=(1,0))
multiplot2.plot(kind='scatter',x='weight',y='horsepower')
plt.axis([1400, 5000, 0, 250])
plt.title('Horsepower Vs. Weight in Europe')
plt.xlabel('Weight')
plt.ylabel('Horsepower')

multiplot3 = plt.subplot2grid(shape=(3,1), loc=(2,0))
multiplot3.plot(kind='scatter',x='weight',y='horsepower')
plt.axis([1400, 5000, 0, 250])
plt.title('Horsepower Vs. Weight in Asia')
plt.xlabel('Weight')
plt.ylabel('Horsepower')
plt.gcf().set_size_inches(8,24)
```
